# COMPUTER VISION


*Utilizando o OpenCV para processamento de imagens*

```python
#Importar biblioteca
import cv2 as cv
from google.colab import files
import matplotlib.pyplot as plt
import numpy as np
```

```python
#caregar imagens 
upload = files.upload()
```

```python
#preparar a imagem
imagem = (r'maringa.jpg')
img_cv = cv.imread(imagem)

print(type(img_cv))
print('tamanho da imagem: ', img_cv.shape)

plt.imshow(img_cv)
```

```python
#passar para RGB
img = cv.cvtColor(img_cv, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray, cmap = plt.cm.gray)
```

```python
#translacao

altura, largura = img.shape[:2]

#deslocamento em X e Y           X         Y
deslocamento = np.float32([[1,0,100],[0,1,100]])
img_trans = cv.warpAffine(img, deslocamento, (largura,altura))
plt.imshow(img_trans)
```

```python
#rotacao

ponto = (largura/2, altura/2) #ponto central
rotacao = cv.getRotationMatrix2D(ponto, 30, 1.0)
img_rot = cv.warpAffine(img,rotacao, (largura, altura))
plt.imshow(img_rot)
```

```python
#inverter a imagem 
#0 - espelha em X
#1 - espelha em y
#-1 espelha em x e y (gira em Z)

espelhar = cv.flip(img, -1)
plt.imshow(espelhar)
```
